```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, color
from skimage.io import imread, imshow

from google.colab import files
uploaded = files.upload()
```

```python
img_bgr = imread('lenna.JPG')

plt.imshow(img_bgr)
plt.show()
```

```python
height, width, _ = img_bgr.shape

for i in range(0, height - 1):
  for j in range(0, width - 1):
  
  # Get the pixel value
    pixel = img_bgr[i, j]

  #negate each channel value
  #substraccing it from 255

  #1st index red pixel
    pixel[0] = 225 - pixel[0]

  #2nd index green pixel
    pixel[1] = 225 - pixel[1]

  #3rd index blue pixel
    pixel[2] = 225 - pixel[2]

  #storage new values in the pixel
    img_bgr[i, j] = pixel
```

```python
# display the native transformed image
plt.imshow(img_bgr)
plt.show()

# histogram plotting of the
#negative transformed image
color = ('b', 'g', 'r')

for i, col in enumerate(color):

  histr = cv2.calcHist([img_bgr], [i], None, [256], [0, 256])

  plt.plot(histr, color = col)
  plt.xlim([0, 256])

plt.show()
```
